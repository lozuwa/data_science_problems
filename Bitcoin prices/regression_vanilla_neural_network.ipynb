## Validation on random data points

```python
features, targets = np.array(features, np.float32), np.array(targets)

# Pick data
random_row = int(np.random.rand()*features.shape[0])
data_val_features, data_val_targets = features[random_row, :], targets[random_row, :]
# Run data
data_val_features = np.array(data_val_features.reshape(1, -1), np.float32)
predicted_targets = np.squeeze(nn.predict(data_val_features))
print(np.sum(predicted_targets - data_val_targets))

plt.subplot(2, 3, 1)
plt.scatter(list(range(data_val_features.shape[1])), data_val_features)
plt.title("Ground truth (Features)")
plt.xlabel("Data points")
plt.ylabel("Weighted_Price")

plt.subplot(2, 3, 3)
plt.scatter(list(range(data_val_targets.shape[0])), data_val_targets)
plt.title("Ground truth (Target)")
plt.xlabel("Data points")
plt.ylabel("Weighted_Price")

plt.subplot(2, 3, 4)
plt.scatter(list(range(data_val_features.shape[1])), data_val_features)
plt.title("Predicted (Features)")
plt.xlabel("Data points")
plt.ylabel("Weighted_Price")

plt.subplot(2, 3, 6)
plt.scatter(list(range(predicted_targets.shape[0])), predicted_targets)
plt.title("Predicted (Target)")
plt.xlabel("Data points")
plt.ylabel("Weighted_Price")

plt.rcParams["figure.figsize"] = (10,8)
plt.show()
```
